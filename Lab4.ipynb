#### 1.	Calculate the interest earned after 5 years on an investment of $2000, assuming an interest rate of 3% compounded annually.

```r
# Interest earned
CI = 2000*(1+(3/100))^5
cat("Total amount:", CI)
cat("\nInterest earned:",CI - 2000)
```
