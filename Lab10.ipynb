<b>1.	Use R-function matrix to create the matrices called A and B: <br>

a)	Take the inverse of A and the transpose of A.  <br>
b)	Multiply A with B. <br>
c)	Estimate the eigenvalues and eigenvectors of A. <br>
d)	For a matrix A, x is an eigenvector, and λ the eigenvalue of a matrix A, if A · x = λ · x. Test it! <br>

```r
A <- matrix(nrow=3, data=c(1,6,-2,2,4,1,3,1,-1))
B <- matrix(nrow = 3, data=c(1,2,-3,4,5,6,7,8,9))
print(A)
print(B)
```

```r
# taking the inverse
print(solve(A))
cat("\n")
# finding the transpose
print(t(A))
```

```r
# Multiply with A and B
A %*% B
```

```r
# Eigen values
ev <- eigen(A)
values <- ev$values
values
```

```r
# Eigen vectors
vectors <- ev$vectors
vectors
```

```r
# A.x
A %*% vectors
```

```r
# λ.X
values * vectors
```

<u>Answer:</u> From the last 2 cells we can see that A·x = λ·x

2.	Create a matrix, called P: <br>
a)	What is the value of the largest eigenvalue (the so-called dominant eigenvalue) and the corresponding eigenvector? <br>
b)	Create a new matrix, T, which equals P, except for the first row, where the elements are 0. <br>
c)	Now estimate N= (I-T)-1, where I is the identity matrix. <br>
</b>

```r
P <- matrix(nrow = 4, data = c(0,0.9775,0,0,0.0043,0.9111,0.0736,0,0.1132,0,0.9534,0.0452,0,0,0,0.9804))
P
```

```r
# largest Eigen value
ev <- eigen(P)
values <- max(ev$values)
values
```

```r
# largest Eigen vector
vectors <- max(ev$vectors)
vectors
```

```r
# creating the new matrix T
T <- P
T[1,] <- 0
T
```

```r
# Estimating (I-T)-1
N <- solve((diag(4)-T)-1)
N
```

<b>4.	Solve the equations 1000 = y ∗ (3 + x) ∗ (1 + y)4 for y and with x varying over the range from 1 to 100.<br> Plot the root as a function of x.
