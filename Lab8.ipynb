<b>1.	Plot a circle by (x,y) points, where x = r · cos(a) and y = r · sin(a), with a the angle, from 0 to 2π, and r the radius.

```r
a <- seq(0,2*pi, length.out = 100)
plot(x= 2*cos(a),y= 2*sin(a))   # here 2 is the radius
```

<b>4.	For the US, the population density in 1900 (N0) was 76.1 million; the population growth can be described with parameter values: a=0.02 yr−1, K = 500 million of people. <br> 
Actual population values are: <br>
1900 1910 1920 1930 1940 1950 1960 1970 1980 <br>
76.1 92.4 106.5 123.1 132.6 152.3 180.7 204.9 226.5 <br>
a)	Plot the population density curve as a thick line, using the US parameter values.<br>
b)	Add the measured population values as points. Finish the graph with titles, labels etc.
</b>

```r
k <- 500
no <- 76.1
a <- 0.02
curve(k/(1+((k-no)-no*exp(-a*(x-1900)))),1900,1980,main="US populatiion",xlab="year",ylab="million",lwd=2)
n <- matrix(ncol=2,data=c(seq(1900,1980,by=10), 76.1,92.4,106.5,123.1,132.6,152.3,180.7,204.9,226.5) )
points(n)
```

<b>5.	Plot curves for mathematical functions with following R-command ”curve”: <br>
>curve(sin(3*pi*x)) <br>
>curve(sin(3*pi*x),from=0,to=2,col="blue", + xlab="x",ylab="f(x)",main="curve")  <br>
>curve(cos(3*pi*x),add=TRUE,col="red",lty=2) >abline(h=0,lty=2) <br>
>legend("bottomleft",c("sin","cos"),text.col=c("blue","red"),lty=1:2)
</b>

```r
par(mfrow=c(2,2))
curve(sin(3*pi*x))
curve(sin(3*pi*x),from = 0,to=2,col="blue",xlab="x",ylab="f(x)",main="curve")

curve(cos(3*pi*x),add=TRUE,col="red",lty=2) 
abline(h=0,lty=2)

legend("bottomleft",c("sin","cos"),text.col=c("blue","red"),lty=1:2)
```
